### Armstrong Number

| Step | Variable / Operation           | Explanation              | Value   |
| ---- | ------------------------------ | ------------------------ | ------- |
| 1    | `n = 153`                      | Function called          | 153     |
| 2    | `s = str(n)`                   | Convert number to string | `"153"` |
| 3    | `power = len(s)`               | Count digits             | `3`     |
| 4    | Loop digits to compute `total` | Armstrong formula        | -       |
| 4.1  | `d = '1'` → `int(d)**3`        | 1³ = 1                   | 1       |
| 4.2  | `d = '5'` → `int(d)**3`        | 5³ = 125                 | 125     |
| 4.3  | `d = '3'` → `int(d)**3`        | 3³ = 27                  | 27      |
| 5    | `total = 1 + 125 + 27`         | Sum of cubes             | `153`   |
| 6    | `total == n`                   | 153 == 153               | True    |
| 7    | Return value                   | Final result             | True    |


```python

def is_armstrong(n):
    s = str(n) #"153"
    power = len(s) # 3
    total = sum(int(d)**power for d in s) # 1^3+ 5^3 + 3^3 =153
    return total == n #153 == 153

# Dry Run
print(is_armstrong(153))
```

### *args & **kwargs

| Step | Operation                                        | Explanation                                                  | Result                                  |
| ---- | ------------------------------------------------ | ------------------------------------------------------------ | --------------------------------------- |
| 1    | Call `demo_args_kwargs(1,2,3, name="A", age=20)` | Three positional arguments, two keyword arguments are passed | -                                       |
| 2    | `args` collects positional arguments             | `args = (1, 2, 3)`                                           | `(1, 2, 3)`                             |
| 3    | `kwargs` collects keyword arguments              | `kwargs = {"name": "A", "age": 20}`                          | `{'name': 'A', 'age': 20}`              |
| 4    | Return `(args, kwargs)`                          | Function returns a tuple of both                             | `((1,2,3), {'name':'A','age':20})`      |
| 5    | Print result                                     | Displays returned value                                      | `((1, 2, 3), {'name': 'A', 'age': 20})` |


```python
def demo_args_kwargs(*args, **Kwargs):
    return args, Kwargs
print(demo_args_kwargs(1,2,3, name="A", age=20))
```

### Closure remembers previous value

| Line | Code            | What Happens      | `self.total` |
| ---- | --------------- | ----------------- | ------------ |
| 1–7  | class Counter   | Class is defined  | -            |
| 8    | `c = Counter()` | `__init__()` runs | 0            |

First call: c(5)
| Step | Code                | Explanation         | `self.total` | Return Value | Output    |
| ---- | ------------------- | ------------------- | ------------ | ------------ | --------- |
| 1    | `c(5)`              | Calls `__call__(5)` | 0 → 5        | -            | -         |
| 2    | `self.total += x`   | 0 + 5               | 5            | -            | -         |
| 3    | `return self.total` | Returns 5           | 5            | 5            | 5 printed |

Second call: c(3)

| Step | Code                | Explanation         | `self.total` | Return Value | Output    |
| ---- | ------------------- | ------------------- | ------------ | ------------ | --------- |
| 1    | `c(3)`              | Calls `__call__(3)` | 5 → 8        | -            | -         |
| 2    | `self.total += x`   | 5 + 3               | 8            | -            | -         |
| 3    | `return self.total` | Returns 8           | 8            | 8            | 8 printed |




```python
class Counter:
    def __init__(self):
        self.total = 0 # 5
    
    def __call__(self, x): #5, #3
        self.total += x #0+5 = 5+3 =8
        return self.total # 8

c = Counter()
print(c(5))   # 5
print(c(3))   # 8
```

### Email validation (manual)

| Step | Operation                         | Value / Result          |
| ---- | --------------------------------- | ----------------------- |
| 1    | Check if "@" and "." in s         | Both present → continue |
| 2    | Find index of "@"                 | at = 3                  |
| 3    | Find last index of "."            | dot = 8                 |
| 4    | Check `at > 0`                    | 3 > 0 → True            |
| 5    | Check `dot > at + 1`              | 8 > 4 → True            |
| 6    | Check `dot < len(s)-1`            | 8 < 11 → True           |
| 7    | All conditions True → return True | Output → True           |


```python
def is_email(s):
    if "@" not in s or "." not in s: return False
    at = s.index("@") #3
    dot = s.rindex(".") #8
    return at>0 and dot>at+1 and dot<len(s)-1

print(is_email("abc@test.com"))
```

### Factorial without recursion

```python
def fac(n): # n=5
    res = 1
    #1,6, i =1,2,3,4,5,6
    for i in range(1, n+1): res*=i  #1*1 =1
    return res
print(fac(5))
```

### Frequency without Counter

| Step | x | d before update    | d after update     |
| ---- | - | ------------------ | ------------------ |
| 1    | 1 | {}                 | {1: 1}             |
| 2    | 2 | {1: 1}             | {1: 1, 2: 1}       |
| 3    | 2 | {1: 1, 2: 1}       | {1: 1, 2: 2}       |
| 4    | 3 | {1: 1, 2: 2}       | {1: 1, 2: 2, 3: 1} |
| 5    | 3 | {1: 1, 2: 2, 3: 1} | {1: 1, 2: 2, 3: 2} |
| 6    | 3 | {1: 1, 2: 2, 3: 2} | {1: 1, 2: 2, 3: 3} |


```python
def freq(lst):
    d = {}#{1:1, 2:2, 3:3}
    for x in lst: # x=1,2,2, 3,3
        d[x] = d.get(x,0)+1 #0+1=1, 0+1=1,1+1 =2, 0+1 = 3
    return d    

print(freq([1,2,2,3,3,3]))    
```

### Even numbers from list

| Step | x | Condition `x%2==0` | Action / Result List |
| ---- | - | ------------------ | -------------------- |
| 1    | 1 | False              | skip                 |
| 2    | 2 | True               | add → [2]            |
| 3    | 3 | False              | skip                 |
| 4    | 4 | True               | add → [2,4]          |
| 5    | 6 | True               | add → [2,4,6]        |


```python
def even_only(lst):
    return [x for x in lst if x%2 == 0]

print(even_only([1,2,3,4,6]))    
```

### Remove HTML tags

| Character | Inside Tag? (before) | Action Taken | Inside Tag? (after) | Output So Far |
| --------- | -------------------- | ------------ | ------------------- | ------------- |
| A         | NO                   | Keep A       | NO                  | "A"           |
| <         | NO                   | Enter tag    | YES                 | "A"           |
| b         | YES                  | Skip b       | YES                 | "A"           |
| >         | YES                  | Exit tag     | NO                  | "A"           |
| B         | NO                   | Keep B       | NO                  | "AB"          |
| <         | NO                   | Enter tag    | YES                 | "AB"          |
| /         | YES                  | Skip /       | YES                 | "AB"          |
| b         | YES                  | Skip b       | YES                 | "AB"          |
| >         | YES                  | Exit tag     | NO                  | "AB"          |
| C         | NO                   | Keep C       | NO                  | "ABC"         |


```python
def strip_html(s):
    out = ""
    inside = False
    for ch in s: # ch = <
        if ch == "<":
            inside = True
        elif ch == ">":
            inside = False
        elif not inside:
            out += ch
    return out

print(strip_html("A<b>B</b>C"))
```

```python
import re

def strip_html(s):
    return re.sub('<.*?>', '', s)

print(strip_html("A<b>B</b>C"))  # Output: ABC
```

### Function returns another function

| Step | Code / Action                               | Explanation / Variables                                |
|------|---------------------------------------------|--------------------------------------------------------|
| 1    | `f = choose("add")`                         | Call `choose` with `"add"`                             |
| 2    | Inside `choose`                             | `op == "add"`, so `return add`                        |
| 3    | `f` now references `add(a,b)` function      | `f` is a function that takes two arguments and adds  |
| 4    | `f(2,3)`                                    | Calls the `add` function with `a=2`, `b=3`           |
| 5    | `return a+b`                                | Computes `2+3 = 5`                                   |
| 6    | Print result                                | `5` is printed                                        |


```python
def choose(op):
    def add(a,b): return a+b
    def mul(a,b): return a*b
    return add if op =="add" else mul

f = choose("add")
print(f(2,3))
```

### Custom sum

total=0  
| Step | x (current value) | total before | total after (`total = total + x`) |
| ---- | ----------------- | ------------ | --------------------------------- |
| 1    | 1                 | 0            | 1                                 |
| 2    | 2                 | 1            | 3                                 |
| 3    | 3                 | 3            | 6                                 |
| 4    | 4                 | 6            | 10                                |


```python
def my_sum(iterable):
    total = 0
    for x in iterable: # x=1,2
        total += x # 0+1=1, 1+2 =3
    return total    
print(my_sum([1,2,3,4]))
```
