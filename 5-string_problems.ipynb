### 1. Reverse each word in a sentence

Input sentence: "hello world"

After split

words = ["hello", "world"]

WORD 1: "hello"

| Step  | Variable    | Value           |
| ----- | ----------- | --------------- |
| Start | w           | "hello"         |
| 1     | ch = "h"    | temp = "h"      |
| 2     | ch = "e"    | temp = "eh"     |
| 3     | ch = "l"    | temp = "leh"    |
| 4     | ch = "l"    | temp = "lleh"   |
| 5     | ch = "o"    | temp = "olleh"  |
| End   | append temp | rev = ["olleh"] |


WORD 2: "world"

| Step  | Variable    | Value                    |
| ----- | ----------- | ------------------------ |
| Start | w           | "world"                  |
| 1     | ch = "w"    | temp = "w"               |
| 2     | ch = "o"    | temp = "ow"              |
| 3     | ch = "r"    | temp = "row"             |
| 4     | ch = "l"    | temp = "lrow"            |
| 5     | ch = "d"    | temp = "dlrow"           |
| End   | append temp | rev = ["olleh", "dlrow"] |


" ".join(rev) → "olleh dlrow"

```python
def reverse_each_word(sentence):
    words = sentence.split() # words = ["hello", "world"]
    rev = []
    for w in words: #w="hello"
        temp = ""
        for ch in w: # ch = 'h'
            temp = ch + temp #'h'+"" =h
        rev.append(temp)   #"olleh"
    return " ".join(rev)      

print(reverse_each_word("hello world"))
```

### 2. Remove all non-alphabetic characters from a string

s = "a1b!"

| Step | ch | Is Alphabet? | Condition Check             | res  |
| ---- | -- | ------------ | --------------------------- | ---- |
| 1    | a  | Yes          | `'a' <= 'a' <= 'z'` → True  | "a"  |
| 2    | 1  | No           | `'a' <= '1' <= 'z'` → False | "a"  |
| 3    | b  | Yes          | `'a' <= 'b' <= 'z'` → True  | "ab" |
| 4    | !  | No           | `'a' <= '!' <= 'z'` → False | "ab" |


```python
def remove_non_alpha(s):
    res = ""
    for ch in s: 
        if ('a'<= ch <= 'z') or ('A' <= ch <= 'z'):
            res += ch
    return res        
print(remove_non_alpha("a1b!"))
```

### 3. Count vowels & consonants

| Step | ch  | ch.isalpha()? | ch in vowels? | v (vowels count) | c (consonants count) | Notes            |
| ---- | --- | ------------- | ------------- | ---------------- | -------------------- | ---------------- |
| 1    | 'a' | True          | True          | 1                | 0                    | 'a' is a vowel   |
| 2    | 'b' | True          | False         | 1                | 1                    | 'b' is consonant |
| 3    | 'C' | True          | False         | 1                | 2                    | 'C' is consonant |


```python
def count_vowels_consonants(s):
    vowels =("aeiouAEIOU")
    v = c = 0
    for ch in s: #ch =a,b
        if ch.isalpha():  # ch =a=b =True
            if ch in vowels: # ch =a=True
                v = +1
            else:
                c += 1
    return v, c                
print(count_vowels_consonants("abC"))
```

### 4. Check if all characters are unique

s= aabc

| Step | Current Character (ch) | Current `seen` Set | Is `ch` already in `seen`? | Action                                                   |
| ---- | ---------------------- | ------------------ | -------------------------- | -------------------------------------------------------- |
| 1    | 'a'                    | {}                 | No                         | Add 'a' → seen = {'a'}                                   |
| 2    | 'a'                    | {'a'}              | **Yes**                    | Duplicate found → return **False** (function stops here) |
| 3    | 'b'                    | —                  | —                          | Not executed (loop stopped)                              |
| 4    | 'c'                    | —                  | —                          | Not executed                                             |



```python
def all_unique(s):
    seen = set() #a
    for ch in s: #ch =a, ch= a, ch=b
        if ch in seen: # a
            return False
        seen.add(ch) # a,b
    return True    
print(all_unique("aabc"))
```

### 5. First non-repeating character

s=aabbcde
| Step | ch  | s.count(ch) | Unique?              |
| ---- | --- | ----------- | -------------------- |
| 1    | 'a' | 2           | No                   |
| 2    | 'a' | 2           | No                   |
| 3    | 'b' | 2           | No                   |
| 4    | 'b' | 2           | No                   |
| 5    | 'c' | 1           | **Yes → return 'c'** |
| 6    | 'd' | —           | Not checked          |
| 7    | 'e' | —           | Not checked          |


```python
def first_non_repeating(s):
    for ch in s: # ch= 'a'
        if s.count(ch) == 1:
            return ch
    return None    
print(first_non_repeating("aabbcde"))
```

### 6. Check anagram without sorting

| Step | ch from a | b_list before             | Found? | b_list after                       |
| ---- | --------- | ------------------------- | ------ | ---------------------------------- |
| 1    | l         | ['s','i','l','e','n','t'] | Yes    | remove 'l' → ['s','i','e','n','t'] |
| 2    | i         | ['s','i','e','n','t']     | Yes    | ['s','e','n','t']                  |
| 3    | s         | ['s','e','n','t']         | Yes    | ['e','n','t']                      |
| 4    | t         | ['e','n','t']             | Yes    | ['e','n']                          |
| 5    | e         | ['e','n']                 | Yes    | ['n']                              |
| 6    | n         | ['n']                     | Yes    | []                                 |


```python
def is_anagram(a, b):
    if len(a) != len(b):
        return False
    
    b_list = list(b) #['s','i','l','e','n','t']

    for ch in a: # ch= l
        if ch in b_list: 
            b_list.remove(ch)
        else:
            return False   
    return True   

print(is_anagram("listen", "silent"))
```

### 8. Convert comma-separated numbers → list of ints

s = "1,2,3,10,20"  

Step 2 – Evaluate: s.split(",") 

s.split(",")  →  ["1", "2", "3", "10", "20"]  

[int(x) for x in ["1", "2", "3", "10", "20"]]

| Iteration | x    | int(x) | Result List So Far |
| --------- | ---- | ------ | ------------------ |
| 1         | "1"  | 1      | [1]                |
| 2         | "2"  | 2      | [1, 2]             |
| 3         | "3"  | 3      | [1, 2, 3]          |
| 4         | "10" | 10     | [1, 2, 3, 10]      |
| 5         | "20" | 20     | [1, 2, 3, 10, 20]  |


```python
def csv_to_int_list(s):
    return [int(x) for x in s.split(",")]

csv_to_int_list("1,2,3,10,20")
```

### 9. Find longest word in sentence

| Step | Variable / Operation | Value / Result            | Explanation                        |
| ---- | -------------------- | ------------------------- | ---------------------------------- |
| 1    | `s.split()`          | `["I", "love", "Python"]` | Extra spaces removed automatically |
| 2    | `longest = ""`       | `""`                      | Start with empty longest word      |


| Iteration | `w`        | `len(w)` | `len(longest)` | Condition (`len(w) > len(longest)`) | Action         | New `longest` |
| --------- | ---------- | -------- | -------------- | ----------------------------------- | -------------- | ------------- |
| 1         | `"I"`      | 1        | 0              | True                                | Update longest | `"I"`         |
| 2         | `"love"`   | 4        | 1              | True                                | Update longest | `"love"`      |
| 3         | `"Python"` | 6        | 4              | True                                | Update longest | `"Python"`    |


```python
def longest_word(s):
    words = s.split()
    longest = ""
    for w in words: # w="I" , w= "love" , w = Python
        if len(w) > len(longest):
            longest = w  # I , love, Python
    return longest      # Python   
print(longest_word("I love Python"))
```

### 10. Count uppercase, lowercase, digits, special chars

Initial: up=0, low=0, dig=0, special=0

| Step | ch  | Condition | Action       | up | low | dig | special |
| ---- | --- | --------- | ------------ | -- | --- | --- | ------- |
| 1    | `a` | lowercase | low += 1     | 0  | 1   | 0   | 0       |
| 2    | `B` | uppercase | up += 1      | 1  | 1   | 0   | 0       |
| 3    | `1` | digit     | dig += 1     | 1  | 1   | 1   | 0       |
| 4    | `!` | special   | special += 1 | 1  | 1   | 1   | 1       |


```python
def count_chars(s):
    up = low = dig = special = 0
    for ch in s:
        if 'A' <= ch <= 'Z': # ch ="B"
            up += 1
        elif 'a' <= ch <= 'z':
            low += 1
        elif '0' <= ch <= '9':
            dig += 1
        else:
            special += 1
    return up, low, dig, special

print(count_chars("aB1!"))
```
