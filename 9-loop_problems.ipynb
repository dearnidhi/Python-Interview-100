### Problem
Print 1â€“5 without using any digit in code.

ord('A') = 65  
Why 65?

Because in the ASCII table,

'A' has value 65   
'B' has value 66   
'C' has value 67  

Loop range = range(65, 65 + 5) â†’ 65 to 69  
Expression printed = i - 65 + 1    
So output becomes: 1, 2, 3, 4, 5 

| Iteration | `i` Value | Calculation (`i - 65 + 1`) | Printed Output |
| --------- | --------- | -------------------------- | -------------- |
| 1         | 65        | 65 - 65 + 1 = **1**        | 1              |
| 2         | 66        | 66 - 65 + 1 = **2**        | 2              |
| 3         | 67        | 67 - 65 + 1 = **3**        | 3              |
| 4         | 68        | 68 - 65 + 1 = **4**        | 4              |
| 5         | 69        | 69 - 65 + 1 = **5**        | 5              |


```python

# Print 1-5 without using any digit
for i in range(ord('A'), ord('A')+5): #range(65,65+5)->65 t0 69 -> i =65,66,67,68,69
    print(i - ord('A')+1) 
    #65-65+1=1
    #66-65+1 =2
    #67 -65+1 =3
```

### Problem
Print star pyramid pattern.

Spaces = (n - i)  
Stars = (2*i - 1)

| i | Spaces (`n-i`) | Stars (`2*i-1`) | Output Line   |
| - | -------------- | --------------- | ------------- |
| 1 | 4              | 1               | `"    *"`     |
| 2 | 3              | 3               | `"   ***"`    |
| 3 | 2              | 5               | `"  *****"`   |
| 4 | 1              | 7               | `" *******"`  |
| 5 | 0              | 9               | `"*********"` |


```python

# Star pyramid pattern
n= 5
for i in range(1, n+1): #range(1,5+1)=>1,2,3,4,5
    print(" "*(n-i) + "*"*(2*i-1)) 
    #" "* 4 + "*" * (2*1-1)= " "*4+"*" *1= " "*4 + "*"= "    *"
    #" "*(3) + "*" *(3)="   ***"
```

### Problem
Print Fibonacci up to n using loop.   
ðŸ‘‰ a = current number   
ðŸ‘‰ b = next number   

Every loop does 2 things:
Print a   

Move forward â†’  
a becomes b   
b becomes a + b  

| Loop # | Printed `a` | New `a = b` | New `b = a + b` |
| ------ | ----------- | ----------- | --------------- |
| 1      | 0           | 1           | 1               |
| 2      | 1           | 1           | 2               |
| 3      | 1           | 2           | 3               |
| 4      | 2           | 3           | 5               |


```python

# Fibonacci up to n
n = 4
a,b = 0,1

for _ in range(n): #4=>i=0,1,2,3
    print(a)
    a,b = b, a+b
```

### Problem
Find max number manually using loop.   
arr = [5,1,9,3,7]

| Step | `x` (Current Value) | Comparison `x > m` | Updated `m` | Explanation                       |
| ---- | ------------------- | ------------------ | ----------- | --------------------------------- |

### Problem
Count frequency of each char manually using loop (no dict).

| Step | `c` | `chars` Before | `freq` Before | Action                         | `chars` After     | `freq` After |
| ---- | --- | -------------- | ------------- | ------------------------------ | ----------------- | ------------ |
| 1    | 'h' | []             | []            | Not in chars â†’ add             | ['h']             | [1]          |
| 2    | 'e' | ['h']          | [1]           | Add                            | ['h','e']         | [1,1]        |
| 3    | 'l' | ['h','e']      | [1,1]         | Add                            | ['h','e','l']     | [1,1,1]      |
| 4    | 'l' | ['h','e','l']  | [1,1,1]       | Exists â†’ index = 2 â†’ freq[2]++ | ['h','e','l']     | [1,1,2]      |
| 5    | 'o' | ['h','e','l']  | [1,1,2]       | Add                            | ['h','e','l','o'] | [1,1,2,1]    |


```python

# Char frequency without dict
s = "hello"
chars = [] 
freq = []
for c in s: # c=h'
    if c not in chars:
        chars.append(c) #'h'
        freq.append(1) #[1]
    else:
        idx = chars.index(c)
        freq[idx] += 1
list(zip(chars,freq))
```

Split a list into even & odd using loop.

| Step | `x` | Condition `x % 2 == 0` | Goes To â†’ | `even` Before | `odd` Before | `even` After | `odd` After |
| ---- | --- | ---------------------- | --------- | ------------- | ------------ | ------------ | ----------- |

| 2    | 2   | 2 % 2 = 0 â†’ âœ”ï¸         | even      | []            | [1]          | [2]          | [1]         |

```python

# Split list into even & odd
arr = [1,2,3,4,5,6]
even, odd = [], []
for x in arr:
    (even if x%2 == 0 else odd).append(x)
even, odd    
```

### Problem
Find longest word manually using loop.

| Variable  | Value                             |
| --------- | --------------------------------- |
| `words`   | `["python", "loops", "practice"]` |
| `longest` | `""` (empty string)               |


| Step | Current Word `w` | `len(w)` | `len(longest)` | Condition `len(w) > len(longest)` | New `longest` |
| ---- | ---------------- | -------- | -------------- | --------------------------------- | ------------- |
| 1    | `"python"`       | 6        | 0              | âœ”ï¸ True                           | `"python"`    |
| 2    | `"loops"`        | 5        | 6              | âŒ False                           | `"python"`    |
| 3    | `"practice"`     | 8        | 6              | âœ”ï¸ True                           | `"practice"`  |


```python

# Find longest word manually
s = "python loops practice"
words = s.split()
longest = "" #0
for w in words: #python, loops , practice
    if len(w) > len(longest): #6>0, 5>6, 8>6
        longest = w #python, practice
longest       
```

### Problem
Find sum of digits of a number using loop.

ðŸ§  Why ord(d) - ord('0')?

ord('0') = 48

ord('1') = 49

â€¦

ord('9') = 57   
ord('9') - ord('0') = 57 - 48 = 9


| Step | `d` (character) | `ord(d)` | `ord('0')` | `ord(d) - ord('0')` | Previous `s` | New `s` |
| ---- | --------------- | -------- | ---------- | ------------------- | ------------ | ------- |
| 1    | `'9'`           | 57       | 48         | 9                   | 0            | 9       |
| 2    | `'8'`           | 56       | 48         | 8                   | 9            | 17      |
| 3    | `'7'`           | 55       | 48         | 7                   | 17           | 24      |


```python

# Sum of digits using loop
num = 987
s = 0
for d in str(num): # d='9'
    s += ord(d) - ord('0') 
    #s= 0+57-48=9
    #s=9+56-48 =17
    #s=17+55-48 =24

print(s)    
```
