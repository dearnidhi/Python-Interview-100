**1) Count frequency of each word in a string using a dictionary.**

text = "hello world hello python world"

| Step | word     | freq.get(word, 0) | New Value Assigned | freq After Step                       |
| ---- | -------- | ----------------- | ------------------ | ------------------------------------- |
| 1    | "hello"  | 0                 | 1                  | {"hello": 1}                          |
| 2    | "world"  | 0                 | 1                  | {"hello": 1, "world": 1}              |
| 3    | "hello"  | 1                 | 2                  | {"hello": 2, "world": 1}              |
| 4    | "python" | 0                 | 1                  | {"hello": 2, "world": 1, "python": 1} |
| 5    | "world"  | 1                 | 2                  | {"hello": 2, "world": 2, "python": 1} |


```python
text = "hello world hello python world"
freq = {}
for word in text.split(): 
    freq[word] = freq.get(word, 0) + 1 #0 +1 =1
print(freq)    
```

**2) Invert a dictionary (keys <-> values).**

d = {"a": 1, "b": 2, "c": 3}

| Step | k (key) | v (value) | Assignment Performed | inverted After Step      |
| ---- | ------- | --------- | -------------------- | ------------------------ |
| 1    | "a"     | 1         | inverted[1] = "a"    | {1: "a"}                 |
| 2    | "b"     | 2         | inverted[2] = "b"    | {1: "a", 2: "b"}         |
| 3    | "c"     | 3         | inverted[3] = "c"    | {1: "a", 2: "b", 3: "c"} |


```python
d = {"a": 1, "b": 2, "c": 3}
inverted = {}
for k, v in d.items():
    inverted[v] = k
print(inverted)    
```

**3) Merge two dictionaries; if key duplicates, sum values.**

Step 1 — Copy d1 into merged

| Iteration | k   | Action          | merged After Step        |
| --------- | --- | --------------- | ------------------------ |
| 1         | "a" | merged["a"] = 3 | {"a": 3}                 |
| 2         | "b" | merged["b"] = 5 | {"a": 3, "b": 5}         |
| 3         | "c" | merged["c"] = 2 | {"a": 3, "b": 5, "c": 2} |


Step 2 — Merge values from d2

| Iteration | k   | merged.get(k, 0) | d2[k] | Operation           | merged After Step                |
| --------- | --- | ---------------- | ----- | ------------------- | -------------------------------- |
| 1         | "b" | 5                | 4     | merged["b"] = 5 + 4 | {"a": 3, "b": 9, "c": 2}         |
| 2         | "c" | 2                | 1     | merged["c"] = 2 + 1 | {"a": 3, "b": 9, "c": 3}         |
| 3         | "d" | 0                | 7     | merged["d"] = 0 + 7 | {"a": 3, "b": 9, "c": 3, "d": 7} |


```python
d1 = {"a": 3, "b": 5, "c": 2}
d2 = {"b": 4, "c": 1, "d": 7}
merged = {}
for k in d1: #k =a,b, c
    merged[k] = d1[k] #{"a":3,"b":5,"c":2}
for k in d2: #k =b, c, d
    merged[k] = merged.get(k,0) + d2[k]    #b =5+4 =9, c=2+1=3, d=0+7=7 =>{ "a":3,b": 9, "c": 3, "d": 7}
print(merged)    

```

**4) Sort dictionary by values (ascending or descending).**

| Step | Operation                          | State / Explanation                                          |
| ---- | ---------------------------------- | ------------------------------------------------------------ |
| 1    | Original dictionary                | `d = {"a": 30, "b": 10, "c": 20}`                            |
| 2    | Convert to items                   | `[("a", 30), ("b", 10), ("c", 20)]`                          |
| 3    | Apply `lambda x: x[1]` for sorting | Extracted values: `30, 10, 20`                               |
| 4    | Sort ascending                     | Sorted list: `[("b", 10), ("c", 20), ("a", 30)]`             |
| 5    | Convert to dict → `asc`            | `asc = {"b": 10, "c": 20, "a": 30}`                          |
| 6    | Sort descending (`reverse=True`)   | Sorted list: `[("a", 30), ("c", 20), ("b", 10)]`             |
| 7    | Convert to dict → `desc`           | `desc = {"a": 30, "c": 20, "b": 10}`                         |
| 8    | Final Output                       | `({'b': 10, 'c': 20, 'a': 30}, {'a': 30, 'c': 20, 'b': 10})` |


```python
d = {"a": 30, "b": 10, "c": 20}
asc = dict(sorted(d.items(), key = lambda x:x[1]))
desc = dict(sorted(d.items(), key = lambda x:x[1], reverse=True))
asc, desc
```

**5) Convert a dict of lists → list of dicts.**

| Step | i | key       | d[key][i] | temp after assignment    | result                                               |
| ---- | - | --------- | --------- | ------------------------ | ---------------------------------------------------- |
| 1    | 0 | "name"    | "A"       | {"name": "A"}            | []                                                   |
| 2    | 0 | "age"     | 20        | {"name": "A", "age": 20} | []                                                   |
| 3    | 0 | loop ends | –         | –                        | [{"name": "A", "age": 20}]                           |
| 4    | 1 | "name"    | "B"       | {"name": "B"}            | [{"name": "A", "age": 20}]                           |
| 5    | 1 | "age"     | 25        | {"name": "B", "age": 25} | [{"name": "A", "age": 20}]                           |
| 6    | 1 | loop ends | –         | –                        | [{"name": "A", "age": 20}, {"name": "B", "age": 25}] |


```python
d = {"name":["A", "B"], "age":[34,78]}
result = []
for i in range(len(d["name"])): # # i =0, i=1
    temp = {}
    for key in d:
        temp[key] = d[key][i] # {"name":"A", "age":34},{"name":B,"age":78}
    result.append(temp)    # {"name":"A", "age":34}, {"name":B,"age":78}
print(result)     
```

**6) Filter dictionary: keep keys with integer values only.**

| Step    | k (key) | v (value) | `isinstance(v, int)` | Added to filtered? | filtered after step  |
| ------- | ------- | --------- | -------------------- | ------------------ | -------------------- |
| Initial | –       | –         | –                    | –                  | `{}`                 |
| 1       | `"a"`   | `10`      | True                 | Yes                | `{"a": 10}`          |
| 2       | `"b"`   | `"hello"` | False                | No                 | `{"a": 10}`          |
| 3       | `"c"`   | `20`      | True                 | Yes                | `{"a": 10, "c": 20}` |

```python

d = {"a": 10, "b": "hello", "c":20, "d":3.5}
filtered = {k:v for k,v in d.items() if isinstance(v, int)}
print(filtered)
```

**7) Remove keys whose value is None, empty string, or empty list.**

| Step    | k (key) | v (value) | Condition `v not in (None, "", [])` | Added to cleaned?     | cleaned After Step     |
| ------- | ------- | --------- | ----------------------------------- | --------------------- | ---------------------- |
| Initial | –       | –         | –                                   | –                     | `{}`                   |
| 1       | `"a"`   | `None`    | False (None is in tuple)            | No                    | `{}`                   |
| 2       | `"b"`   | `""`      | False ("" is in tuple)              | No                    | `{}`                   |
| 3       | `"c"`   | `[]`      | False (`[]` is in tuple)            | No                    | `{}`                   |
| 4       | `"d"`   | `10`      | True                                | Yes → add `"d": 10`   | `{"d": 10}`            |
| 5       | `"e"`   | `"ok"`    | True                                | Yes → add `"e": "ok"` | `{"d": 10, "e": "ok"}` |


```python
d = {"a": None, "b": "", "c":[], "d":10, "e":"ok"}
cleaned = {}
for k, v in d.items():
    if v not in(None, "",[]) and v !=[]:
        cleaned[k] = v
print(cleaned)        
```

**8) Given a sentence, build a dict mapping each word → its length.**

| Step    | w (current word) | len(w) | Action Performed    | word_len After Update                               |
| ------- | ---------------- | ------ | ------------------- | --------------------------------------------------- |
| Initial | –                | –      | Initialize          | `{}`                                                |
| 1       | `"Python"`       | `6`    | Add `"Python": 6`   | `{"Python": 6}`                                     |
| 2       | `"is"`           | `2`    | Add `"is": 2`       | `{"Python": 6, "is": 2}`                            |
| 3       | `"super"`        | `5`    | Add `"super": 5`    | `{"Python": 6, "is": 2, "super": 5}`                |
| 4       | `"powerful"`     | `8`    | Add `"powerful": 8` | `{"Python": 6, "is": 2, "super": 5, "powerful": 8}` |


```python
sentence = "Python is super powerful"
word_len = {}
for w in sentence.split():# w =["Python","is", "super", " powerful"]
    word_len[w] = len(w) #{"Python":6, "is":2, "super":5, "powerful": 8}
word_len    
```

**9) Given a list of dicts, remove duplicate dicts.**

| Step    | d (current dict) | sorted(d.items())   | t (tuple form)      | `t in seen?` | Action               | seen after                                 | unique after                     |
| ------- | ---------------- | ------------------- | ------------------- | ------------ | -------------------- | ------------------------------------------ | -------------------------------- |
| Initial | –                | –                   | –                   | –            | Initialize           | `[]`                                       | `[]`                             |
| 1       | `{"a":1,"b":2}`  | `[("a",1),("b",2)]` | `(("a",1),("b",2))` | No           | Add to seen & unique | `[ (("a",1),("b",2)) ]`                    | `[{"a":1,"b":2}]`                |
| 2       | `{"a":1,"b":2}`  | `[("a",1),("b",2)]` | `(("a",1),("b",2))` | Yes          | Skip                 | `[ (("a",1),("b",2)) ]`                    | `[{"a":1,"b":2}]`                |
| 3       | `{"a":2,"b":3}`  | `[("a",2),("b",3)]` | `(("a",2),("b",3))` | No           | Add to seen & unique | `[ (("a",1),("b",2)), (("a",2),("b",3)) ]` | `[{"a":1,"b":2}, {"a":2,"b":3}]` |


```python

data =[
    {"a":1,"b":2},
    {"a":1,"b":2},
    {"a":2,"b":3}
    ]

seen = []
unique =[]
for d in data:
    t = tuple(sorted(d.items()))
    if t not in seen:
        seen.append(t)
        unique.append(d)
unique
```

**10) Convert two lists into a dictionary without using zip().**

| Step    | i | keys[i] | values[i] | Update to result | result After Step             |
| ------- | - | ------- | --------- | ---------------- | ----------------------------- |
| Initial | – | –       | –         | result = {}      | `{}`                          |
| 1       | 0 | "a"     | 10        | result["a"] = 10 | `{"a": 10}`                   |
| 2       | 1 | "b"     | 20        | result["b"] = 20 | `{"a": 10, "b": 20}`          |
| 3       | 2 | "c"     | 30        | result["c"] = 30 | `{"a": 10, "b": 20, "c": 30}` |


```python

keys = ["a", "b", "c"]
values = [10,20,30]
result = {}
for i in range(len(keys)): #3=> i =0,1,2
    result[keys[i]] = values[i] #keys[0] = values[0]=>{"a":10, "b":20 , "c":30}
result    
```
